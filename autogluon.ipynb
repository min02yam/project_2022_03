```python
import warnings
warnings.filterwarnings(action='ignore') 

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
train = pd.read_csv("/content/drive/MyDrive/2022_03/data/train.csv")
test = pd.read_csv("/content/drive/MyDrive/2022_03/data/test.csv")
submission = pd.read_csv("/content/drive/MyDrive/2022_03/data/sample_submission.csv")
```

```python
feature = train.drop(['predicted_weight_g'],axis=1).columns
target = 'predicted_weight_g'
```

```python
from autogluon.tabular import TabularPredictor
predictor = TabularPredictor(label = target).fit(train_data=train)
```

```python
predictor.fit_summary()
```

```python
pred = predictor.predict(test[feature]) 
submission['predicted_weight_g'] = pred
submission['case'] = test['case']

submission.to_csv('autogluon_base.csv', index=False)
submission.head()
```
